```
import  tensorflow as tf 
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```
#dowload data set from 
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```
#define class
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```
def lookup(type_class):
  index = 0
  for key,value in enumerate(train_labels):
    if(type_class == value):
      index = key
      break
  return index
```

```
def showImageByType(type_class):
  #test to visualization
  i = lookup(type_class) 
  plt.figure()
  plt.imshow(train_images[i])
  plt.colorbar()
  plt.grid(False)
  plt.show()

class_names
```

```
def ask_show_product():
  print("============= SHOP ==========")
  print("= what product you want to look at?")
  print("= 0.T-shirt/top ")
  print("= 1.Trouser ")
  print("= 2.Pullover")
  print("= 3.Dress ")
  print("= 4.Coat ")
  print("= 5.Sandal")
  print("= 6.Shirt ")
  print("= 7.Sneaker ")
  print("= 8.Bag ")
  print("= 9.Ankle boot ")
  order = int(input("order? :"))
  if(order < 10 and order >= 0):
    showImageByType(order)
  else:
    print("Not Found Product")

```

```
ask_show_product()
```

```
def show_random_data(count=25):
  plt.figure(figsize=(15,15))
  for i in range(count):
      plt.subplot((count/2)+1,(count/2),i+1)
      plt.xticks([])
      plt.yticks([])
      plt.grid(False)
      plt.imshow(train_images[i], cmap=plt.cm.binary)
      plt.colorbar()
      plt.xlabel(class_names[train_labels[i]])
  plt.show()
```
